```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import numpy as np
```

```python
#dataset's path 
dir_path_DataSet = 'drive/My Drive/medical_image_recognition/datasets/'

#fold name for saving files(kmean/PCA)
vlad_fold="vlad_fold/"
#Y's fold 
y_fold='datanp_vgg/'

# docs=['chest_xray','miniMIT_Etus']#,'chest_xray','kvasir-dataset-v2']#,'miniMIT_Etus','kvasir-dataset-v2']
docs=['miniMIT_Etus']
layers=['block5_pool']#,'block8_9_conv']#,'block8_8_conv']
scales = [0,1,2]
```

```python
for doc in docs:
    print(doc)
    for layer in layers:
        for n in scales:
            print(layer)
            v_train_64 = np.load(dir_path_DataSet+doc+'/'+vlad_fold+doc+'_'+layer+'_'+'n'+str(n)+'_vlad_train(64).npy')
            v_train_256= np.load(dir_path_DataSet+doc+'/'+vlad_fold+doc+'_'+layer+'_'+'n'+str(n)+'_vlad_train(256).npy')
            v_test_64  = np.load(dir_path_DataSet+doc+'/'+vlad_fold+doc+'_'+layer+'_'+'n'+str(n)+'_vlad_test(64).npy')
            v_test_256 = np.load(dir_path_DataSet+doc+'/'+vlad_fold+doc+'_'+layer+'_'+'n'+str(n)+'_vlad_test(256).npy')

            print(v_train_64.shape)
            y_train=np.load(dir_path_DataSet + doc+'/'+ y_fold + doc +'Y_train_'+layer + '(' + str(0) + ').npy')
            y_test=np.load( dir_path_DataSet + doc+'/'+y_fold + doc +'Y_test_'+layer + '(' + str(0) + ').npy')
            print(v_train_256.shape)


            #################################################
            print('SUPPORT VECTOR MACHINE 256')

            parameters = {'C':[0.0001,0.001,0.01, 0.1, 1, 10,100]}
            clf = SVC(kernel='linear')
            clf = GridSearchCV(clf, parameters, cv=3)
            clf.fit(v_train_256, y_train)

            print('Accuracy vlad 256: %.2f' % clf.score(v_test_256, y_test))

            print('prediction SVM [%s,%s]' %(doc , layer))

            ####################################################

            print('SUPPORT VECTOR MACHINE 64')

            parameters = {'C':[0.0001,0.001,0.01, 0.1, 1, 10,100]}
            clf = SVC(kernel='linear')
            clf = GridSearchCV(clf, parameters, cv=3)
            clf.fit(v_train_64, y_train)

            print('Accuracy vlad 64: %.2f' % clf.score(v_test_64, y_test))
        
        
        
```
