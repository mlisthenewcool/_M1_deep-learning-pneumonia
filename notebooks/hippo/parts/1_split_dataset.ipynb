- dataset_name
    - [class 1, ..., class n]
        - [image 1, ..., image m]
 
## to 

 - dataset_name
    - [train, val, test] (test is optional)
        - [class 1, ..., class n]
            - [image 1, ..., image m]

```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue May 14 18:06:30 2019

@author: hippolyte
"""

import split_folders

ROOT_PATH = 'drive/My Drive/master1/medical_image_recognition/datasets/'
INPUT_PATH = ROOT_PATH + 'kvasir_v2/'
OUTPUT_PATH = ROOT_PATH + 'kvasir_v2_split/'

TRAIN_RATIO = 0.6
TEST_RATIO = 0.2
VAL_RATIO = 0.2

SEED = 1234
```

```
split_folders.ratio(INPUT_PATH,
                    output=OUTPUT_PATH,
                    seed=SEED,
                    ratio=(TRAIN_RATIO, VAL_RATIO, TEST_RATIO))
```
